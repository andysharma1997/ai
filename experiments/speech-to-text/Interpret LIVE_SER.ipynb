```python
with open(file) as f:
    lines = f.readlines()
```

```python
import wave
import contextlib
def get_duration(fname):
    with contextlib.closing(wave.open(fname,'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        duration = frames / float(rate)
        return duration
        
```

```python
import numpy as np
mode = ['low', 'medium', 'high']
from_time = 0.0
import librosa
duration = get_duration('/home/absin/Downloads/abc.wav')
chunk_time = duration/len(lines)
to_time = chunk_time
for line in lines:
    
    ss = line.split('\t')
    v = [ss[1],ss[2],ss[3]]
    a = [ss[4],ss[5],ss[6]]
    #print('{} --> {}'.format(max(v), max(a)))
    #print('{} --> {}'.format(np.argmax(v), np.argmax(a)))
    fff = float(max(a).strip().replace('.', ''))
    if fff == -1:
        print('{} - {} ---> Neutral'.format(from_time, to_time))
    else:
        print('{} - {} -> Valence: {} Arousal: {} - {}'.format(from_time, to_time, mode[np.argmax(v)], mode[np.argmax(a)], max(a)))
    from_time+=chunk_time
    to_time+=chunk_time
```
